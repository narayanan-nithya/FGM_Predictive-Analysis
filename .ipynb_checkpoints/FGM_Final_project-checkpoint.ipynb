```python
 %matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("FGM_All.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
df
```

```python
# Set features. This will also be used as your x values.

selected_features = df[['Prevarence_G_W','Urban_G_W', 'Rural_G_W', 'Low_Income_G_W', 'Middle_income_G_W', 'High_Income_G_W', 'Oppose_Cont_G_W','Oppose_Cont_B_M','Upper2ndary_Educ_Female','Upper2ndary_Educ_Male']]
selected_features
```

```python
# Drop the null rows
selected_features1 = selected_features .dropna()
selected_features1
```

```python
X=selected_features1[['Prevarence_G_W','Urban_G_W', 'Rural_G_W', 'Low_Income_G_W', 'Middle_income_G_W', 'High_Income_G_W', 'Oppose_Cont_G_W','Oppose_Cont_B_M','Upper2ndary_Educ_Female','Upper2ndary_Educ_Male']]
y=selected_features1['Prevarence_G_W']
y
```

```python
# Group y in two categories. 1 is higher prevalence (is splitting at 30%, it seems to create the 2 groups that are about the same size
selected_features1['Prevarence_G_W1']=[1 if x >=30 else 0 for x in selected_features1['Prevarence_G_W']] 
print(selected_features1)
```

```python
# recreate X and y based on selected_features1

X=selected_features1[['Urban_G_W', 'Rural_G_W', 'Low_Income_G_W', 'Middle_income_G_W', 'High_Income_G_W', 'Oppose_Cont_G_W','Oppose_Cont_B_M','Upper2ndary_Educ_Female','Upper2ndary_Educ_Male']]
y=selected_features1['Prevarence_G_W1']
y
```

```python
X=selected_features1[['Urban_G_W', 'Rural_G_W', 'Low_Income_G_W', 'Middle_income_G_W', 'High_Income_G_W', 'Oppose_Cont_G_W','Oppose_Cont_B_M','Upper2ndary_Educ_Female','Upper2ndary_Educ_Male']]
y = selected_features1["Prevarence_G_W1"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
print(f"Training Data Score : {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```
